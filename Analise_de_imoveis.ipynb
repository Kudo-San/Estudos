# Conhecendo a base de dados

## Importando os dados


```python
# Importando a Lib.
import pandas as pd
```

```python
# Carregando a base de dados
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'

# Lendo o arquivo CSV diretamente da URL 
pd.read_csv(url)
```

Obs: A estruturação de uma tabela, não está bem formatada, pois os dados constam em uma só coluna, separados por ponto e vírgula. Arquivos do tipo CSV têm a vírgula como separador.

A função read_csv() considera, por padrão, que o arquivo lido tem a vírgula como separador. Quando este não é o caso, é necessário especificar, na função, que o arquivo possui um separador diferente:

```python
# Lendo o arquivo CSV diretamente da URL com separador correto.
pd.read_csv(url, sep=';')
```

```python
# Atribuindo o DataFrame a uma variável
dados = pd.read_csv(url, sep=';')

# Visualizando as 10 primeiras linhas do DataFrame
dados.head(10)
```

# Análise exploratória de dados
 Qual o valor médio de aluguel por tipo de imóvel?

```python
# Visualizando as 5 primeiras linhas do DataFrame
dados.head(5)
```

```python
# Metodo mean para calcular a média de um conjunto de dados.
dados['Valor'].mean()
```

```python
# nemeric_only=True especifica onde a média será calculada
dados.groupby('Tipo').mean(numeric_only=True)
```

Ao executar essa célula acima, estou agrupando o dataframe dados conforme a coluna Tipo e calculando a média de valores conforme esse agrupamento. visualizando o resultado!

```python
# Melhorando a visualização do dataframe
dados.groupby('Tipo')['Valor'].mean()
```

```python
# Transformando a visualização em um Dataframe
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
# Salvando em uma variavel (df_preco_tipo)
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

#Plotando o gráfico utilizando a variavel criada.
df_preco_tipo.plot(kind='barh', figsize=(14, 10), color = 'green');
```

## Removendo os imóveis comerciais

```python
# Para visualizar os tipos únicos presentes nessa coluna, usarei o método .unique().
dados.Tipo.unique()
```

O método .unique() retornou uma lista com todos os valores únicos existentes na coluna "Tipo". Ou seja, todos os tipos de imóveis que possuí na base de dados.

```python
# Criando uma variavel com nome de (imoveis_comerciais), do tipo lista e add somente o que não vou utilizar.
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
                                        
```

```python
# Uteilizarei o metodo query, vai filtrar o que é relevante para analise.
dados.query('@imoveis_comerciais in Tipo')
```

```python
# Novamente utilizando o metodo query, porem com NOT IN para filtrar somente o que preciso.
dados.query('@imoveis_comerciais not in Tipo')
```

1. Na saída acima temos na coluna "Tipo", apenas imóveis do tipo residencial como quitinete, casa e apartamento. 
2. Quantidade de linhas: antes 9800 linhas, correspondentes aos imóveis que foram removido da base de dados.
3. Agora, temos em torno de 23 mil.

```python
# Criando a variavel DF para salvar o que foi modificado.
df = dados.query('@imoveis_comerciais not in Tipo')
df.head(5)
```

```python
# Verificando a nova lista que foi modificada, utilizando novamente o metodo .unique
df.Tipo.unique()
```

```python
# Salvando em uma nova variavel (df_preco_tipo)
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

#Plotando o gráfico utilizando a variavel criada:
df_preco_tipo.plot(kind='barh', figsize=(14, 10), color = 'orange');
```

## Filtros

1. Apartamentos que possuem `1 quarto` e `aluguel menor que 1200`**

```python
# Aplicando filtros apartamentos que contem 1 quarto
df['Quartos'] == 1
```

```python
# Amarzenando em uma variavel
selecao1 = df['Quartos'] == 1

# Visualizando o resultado
df[selecao1]
```

```python
# Criando uma variavel (selecao2) com filtro na tabela Valor <1200
selecao2 = df['Valor'] < 1200

# visualizando o resultado.
df[selecao2]
```

```python
# Juntando os resultados das variaveis (selecao1 e selecao2) em uma nova variavel (selecao_final) 
selecao_final = (selecao1) & (selecao2)

# Visualizando o resultado
df[selecao_final]
```

```python
# Salvando df1
df1 = df[selecao_final]
```

### **2. `Apartamentos` que possuem pelo menos `2 quartos`, `aluguel menor que 3000` e `area maior que 70`**

```python
# Criando uma variavel (selecao) para filtrar na tabela quartos qtd. até 2
# (&=and) juntar com proximo filtro da tabela Valor menor que 3000
# (&=and) juntar com proximo filtro da tabela area maior que 70
selecao = (df['Quartos'] >= 2) & (df['Valor'] < 3000) & (df['Area'] > 70)

# O resultado em Dataframe
df[selecao]
```

```python
# Salvando o filtro df2
df2 = df[selecao]
```

# Manipulando os dados
 Criando colunas numéricas

```python
 # Carregando a base de dados
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'

dados = pd.read_csv(url, sep=';')

dados.head()
                    
```

```python
# Criando uma variavel (valor_por_mes) como uma nova coluna, para visualizar os valores do mês.
# Juntando os dados da tabela valor + condominio
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']

dados.head()
```

```python
# Criando uma variavel (valor_por_ano) como uma nova coluna, para visualizar os valores do ano.
# Juntando os dados da tabela valor + condominio
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']

dados.head()
```

## Criando colunas categóricas

```python
# Criando uma coluna (descrição) e somando strings
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']

dados.head()
```

```python
# Criando uma coluna de descrição e adicionando as informações.

dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                     dados['Quartos'].astype(str)  + ' quartos(s) ' + \
                     ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem. '
dados.head()
```
